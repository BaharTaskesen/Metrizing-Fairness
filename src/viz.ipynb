```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import rcParams

"""This code prepares Figure A.1"""
```

```python
stdvs = pd.DataFrame([[0.041, 0.072, 0.015, 0.016], 
                      [0.014, 0.012, 0.014, 0.012],
                      [0.019, 0.011, 0.011, 0.018],
                      [0.0, 0.081, float('Nan'), 0.002]])
stdvs.columns = ['Zafar et al.', 'Cho et al.', 'Oneto et al.', 'MFL (Ours)']
stdvs.index = ['Drug', 'Communities&Crime', 'Compas', 'Adult']
stdvs = stdvs.T
```

```python
means = pd.DataFrame([[0.67, 0.77, 0.78, 0.83], 
                      [0.79, 0.81, 0.71, 0.83],
                      [0.625, 0.667, 0.621, 0.666],
                      [0.8, 0.74, float('Nan'), 0.84]])
means.columns = ['Zafar et al.', 'Cho et al.', 'Oneto et al.', 'MFL (Ours)']
means.index = ['Drug', 'Communities&Crime', 'Compas', 'Adult']
means = means.T
```

```python
means
```

```python
idxs = np.array([0,1,2,3])
shift=4
plt.figure(figsize=(16,4))

color_dict = {
    'Zafar et al.' : 'C0',
    'Cho et al.' : 'C1', 
    'Oneto et al.': 'C2', 
    'MFL (Ours)' : 'C3'
}
marker_size=8
error_size=4

for method, idx in zip(['Zafar et al.', 'Cho et al.', 'Oneto et al.', 'MFL (Ours)'], idxs):
    plt.errorbar(idx, means['Drug'][method], yerr=stdvs['Drug'][method], fmt='s', color=color_dict[method],
                 ecolor='lightgray', elinewidth=error_size, capsize=0, label=method, markersize=marker_size)

    plt.errorbar(idx+shift+4, means['Communities&Crime'][method], yerr=stdvs['Communities&Crime'][method], fmt='s', 
                 color=color_dict[method], ecolor='lightgray', elinewidth=error_size, capsize=0, markersize=marker_size)

    plt.errorbar(idx+2*shift+8, means['Compas'][method], yerr=stdvs['Compas'][method], fmt='s',
                 color=color_dict[method], ecolor='lightgray', elinewidth=error_size, capsize=0, markersize=marker_size)

    plt.errorbar(idx+3*shift+12, means['Adult'][method], yerr=stdvs['Adult'][method], fmt='s',
                 color=color_dict[method], ecolor='lightgray', elinewidth=error_size, capsize=0, markersize=marker_size)
    
plt.legend(loc='lower right')
plt.xlim([-1,30])
plt.xticks([1.5, 9.5, 17.5, 25.5], means.columns)
plt.show()

```
